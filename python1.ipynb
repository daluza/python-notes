```python
def getSum(myList):
  sum = 0
  for item in myList:
    sum = sum + item
    print(sum)
  return sum
```

```python
#Quadratic time complexity. O(n2)
def getSumQ(myList):
  sum = 0
  for row in myList:
    for item in row:
      sum += item
  return sum
```

```python
#Logarithmic time complexity. O(log n)
def searchBinary(myList, item):
  first = 0
  last = len(myList)-1
  foundFlag = False
  while( first <= last and not foundFlag):
    mid = (first + last)//2
    if myList[mid] == item :
      foundFlag = True
    else:
      if item < myList[mid]:
        last = mid -1
      else:
        first = mid + 1
  return foundFlag
```

```python
#Abstract Data Types (TDA)
#Stacks - FILO
class Stack:
  def __init__(self):
    self.items = []
  def push(self, item):
    self.items.append(item)
  def isEmpty(self):
    return self.items == []
  def pop (self):
    return self.items.pop()
  def peek(self):
    return self.items[len(self.items)-1]
  def size(self):
    return len(self.items)
```

```python
#Queues - FIFO
class Queue():
  def __init__(self):
    self.items = []
  def isEmpty(self):
    return self.items == []
  def enqueue(self, item):
    self.items.insert(0,item)
  def dequeue(self):
    return self.items.pop()
  def size(self):
    return len(self.items)    
  def peek(self):
    return self.items[len(self.items)-1]
```
